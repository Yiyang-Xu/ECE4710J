```python
import math

def iteration(theta0_t, theta1_t):
    theta0_itr = theta0_t + theta1_t + 0.5
    theta1_itr = theta0_t + theta1_t + 2*math.cos(theta1_t)
    print(theta0_itr, theta1_itr)
    return theta0_itr, theta1_itr
    
theta0_t = 0
theta1_t = 0
for i in range(1,100):
    theta0_t, theta1_t = iteration(theta0_t, theta1_t)
```
